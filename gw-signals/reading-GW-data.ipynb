Then, can read in files as follows (you can find the example file [here](https://github.com/gwpy/gwpy/tree/main/gwpy/testing/data))

```python
from gwpy.timeseries import TimeSeries
data = TimeSeries.read('../data/HLV-HW100916-968654552-1.gwf', 'L1:LDAS-STRAIN')
print(data[1:10])
```

```python
times = data.times 
strain = data.value
plt.plot(times, strain)
plt.show()
```

Some more comments about reading in:
- you can give multiple file names (multiple channels) as a list as first (second) argument
- you can use the `nproc` keyword argument to specify the number of cores to speed up reading in a large number of files.
- example can be found [here](https://gwpy.github.io/docs/stable/timeseries/io/#gwf-library-availability)
